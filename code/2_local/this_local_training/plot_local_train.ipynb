```python
import numpy as np

# Data
hospitals = ['Hospital 55', 'Hospital 43', 'Hospital 18']
models = ['VGG-19', 'ResNet-50', 'ResNet-101', 'DenseNet-161']

data_55 = {
    'Total Acc': [0.296, 0.44, 0.396, 0.572],
    'NFF Acc': [0.16, 0.359, 0.304, 0.529],
    'AFF Acc': [0.877, 0.787, 0.787, 0.755],
    'Precision': [0.197, 0.224, 0.21, 0.274],
    'Recall': [0.877, 0.787, 0.787, 0.755],
    'AUC': [0.518, 0.573, 0.545, 0.642],
    'MCC': [0.04, 0.121, 0.079, 0.223],
    'F1 Score': [0.322, 0.348, 0.331, 0.402],
    'PRC-AUC': [0.196, 0.217, 0.206, 0.253]
}

data_43 = {
    'Total Acc': [0.64, 0.799, 0.78, 0.73],
    'NFF Acc': [0.604, 0.879, 0.847, 0.707],
    'AFF Acc': [0.794, 0.457, 0.496, 0.825],
    'Precision': [0.32, 0.47, 0.433, 0.399],
    'Recall': [0.794, 0.457, 0.496, 0.825],
    'AUC': [0.699, 0.668, 0.672, 0.766],
    'MCC': [0.313, 0.339, 0.326, 0.428],
    'F1 Score': [0.456, 0.463, 0.462, 0.538],
    'PRC-AUC': [0.293, 0.318, 0.311, 0.362]
}

data_18 = {
    'Total Acc': [0.64, 0.799, 0.78, 0.73],
    'NFF Acc': [0.604, 0.879, 0.847, 0.707],
    'AFF Acc': [0.794, 0.457, 0.496, 0.825],
    'Precision': [0.32, 0.47, 0.433, 0.399],
    'Recall': [0.794, 0.457, 0.496, 0.825],
    'AUC': [0.699, 0.668, 0.672, 0.766],
    'MCC': [0.313, 0.339, 0.326, 0.428],
    'F1 Score': [0.456, 0.463, 0.462, 0.538],
    'PRC-AUC': [0.293, 0.318, 0.311, 0.362]
}

# Find the best performance for each metric and model
best_performances = {}
for hospital_data, hospital_name in zip([data_55, data_43, data_18], hospitals):
    best_performances[hospital_name] = {}
    for metric, values in hospital_data.items():
        best_idx = np.argmax(values)
        best_model = models[best_idx]
        best_value = values[best_idx]
        best_performances[hospital_name][metric] = {'Model': best_model, 'Value': best_value}

# Generate LaTeX table
latex_table = "\\begin{table}[h]\n"
latex_table += "    \\centering\n"
latex_table += "    \\begin{tabular}{lcc}\n"
latex_table += "    \\toprule\n"
latex_table += "    \\textbf{Hospital} & \\textbf{Best Model} & \\textbf{Best Value} \\\\\n"
latex_table += "    \\midrule\n"
for hospital_name, metrics in best_performances.items():
    for metric, info in metrics.items():
        latex_table += f"    {hospital_name} {metric} & {info['Model']} & {info['Value']:.3f} \\\\\n"
latex_table += "    \\bottomrule\n"
latex_table += "    \\end{tabular}\n"
latex_table += "    \\caption{Best Performances Across Hospitals}\n"
latex_table += "    \\label{tab:best_performances}\n"
latex_table += "\\end{table}\n"

print(latex_table)
```
