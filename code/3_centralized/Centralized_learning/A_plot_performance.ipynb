```python
import pandas as pd

# 读取CSV文件
file_path = '/local/data1/honzh073/local_repository/FL/code/6_model_test/hospital55&33.csv'  # 替换为你的CSV文件路径
df = pd.read_csv(file_path)

# 统计医院数量
num_hospitals = df['HospitalID'].nunique()

# 统计AFF患者数量和NFF患者数量
aff_patients = df[df['Label'] == 'AFF']['PatientID'].nunique()
nff_patients = df[df['Label'] == 'NFF']['PatientID'].nunique()

# 统计AFF label的数量和NFF label的数量
aff_labels = df[df['Label'] == 'AFF'].shape[0]
nff_labels = df[df['Label'] == 'NFF'].shape[0]

# 计算AFF患者和NFF患者的比例
aff_patient_ratio = aff_patients / (aff_patients + nff_patients)
nff_patient_ratio = nff_patients / (aff_patients + nff_patients)

# 计算AFF label和NFF label的比例
aff_label_ratio = aff_labels / (aff_labels + nff_labels)
nff_label_ratio = nff_labels / (aff_labels + nff_labels)

# 打印结果
print(f'医院数量: {num_hospitals}')
print(f'AFF患者数量: {aff_patients}')
print(f'NFF患者数量: {nff_patients}')
print(f'AFF label数量: {aff_labels}')
print(f'NFF label数量: {nff_labels}')
print(f'AFF患者比例: {aff_patient_ratio:.2f}')
print(f'NFF患者比例: {nff_patient_ratio:.2f}')
print(f'AFF label比例: {aff_label_ratio:.2f}')
print(f'NFF label比例: {nff_label_ratio:.2f}')
```
