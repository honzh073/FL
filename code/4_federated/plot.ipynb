```python
import matplotlib.pyplot as plt

# 从文件读取数据
log_file_path = '/local/data1/honzh/data/suzuki/3dd1be43-2b33-4d68-baff-895916b3e962/log.txt'
values = {'train_loss': [], 'train_accuracy': [], 'train_MCC': [], 'val_acc_global_model': [], 'val_acc_global_model': []}

with open(log_file_path, 'r') as file:
    for line in file:
        if "train_loss:" in line:
            values['train_loss'].append(float(line.split(":")[1].strip()))
        elif "train_accuracy:" in line:
            values['train_accuracy'].append(float(line.split(":")[1].strip()))
        elif "train_MCC:" in line:
            values['train_MCC'].append(float(line.split(":")[1].strip()))
        elif "val_acc_global_model:" in line:
            values['val_acc_global_model'].append(float(line.split(":")[1].strip()))

# 绘制图表
epochs = range(1, len(values['train_loss']) + 1)

plt.figure(figsize=(12, 8))

# 绘制训练损失
plt.subplot(3, 1, 1)
plt.plot(epochs, values['train_loss'], label='Training Loss')
plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# 绘制训练准确度
plt.subplot(3, 1, 2)
plt.plot(epochs, values['train_accuracy'], label='Training Accuracy')
plt.title('Training Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# 绘制训练MCC
plt.subplot(3, 1, 3)
plt.plot(epochs, values['train_MCC'], label='Training MCC')
plt.title('Training MCC')
plt.xlabel('Epochs')
plt.ylabel('MCC')
plt.legend()

plt.tight_layout()
plt.suptitle('Training Metrics over Epochs', fontsize=16)
plt.show()
```
