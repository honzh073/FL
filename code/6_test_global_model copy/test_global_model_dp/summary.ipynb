```python
import numpy as np
import matplotlib.pyplot as plt

# 数据
models = ["VGG 19", "resnet 50", "resnet 101", "resnet 152", "densenet 161"]
hospitals_avg_acc = np.array([
    [88.3587279467067, 81.53391008341211, 82.83887729883267],
    [90.43218976832975, 90.49157423709335, 90.16431050322272],
    [92.18680581033118, 90.60239730502543, 89.73977058367468],
    [91.20732514386049, 90.89114450794729, 91.01168763048115],
    [83.57654970890194, 83.89978678055978, 84.2723406076776]
])

# 计算平均值和方差
avg_acc_mean = np.mean(hospitals_avg_acc, axis=1)
avg_acc_std = np.std(hospitals_avg_acc, axis=1)

# 绘制柱状图
fig, ax = plt.subplots()
x = np.arange(len(models))
bar_width = 0.3

bars = ax.bar(x, avg_acc_mean, bar_width, label='Avg ACC', yerr=avg_acc_std, capsize=5)

# 添加数值标签
for i, bar in enumerate(bars):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 1.2, f'{avg_acc_mean[i]:.2f}', ha='center', va='bottom')

# 添加标签和标题
ax.set_xlabel('Models')
ax.set_ylabel('Accuracy (%)')
ax.set_title('Average Accuracy Across Hospitals for Different Models')
ax.set_xticks(x)
ax.set_xticklabels(models)
# ax.legend()

# 显示柱状图
plt.show()
```
