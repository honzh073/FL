```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Updated data for 100 times average
data = {
    'Model': ['VGG 19', 'ResNet 50', 'ResNet 101', 'DenseNet 161', 'Efficient B2'],
    'Total Accuracy (Mean)': [62.1368, 71.6119, 77.0319, 72.7368, 79.9452],
    'Total Accuracy (Std Dev)': [1.6501, 1.4409, 2.0639, 1.4057, 1.9114],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the style for better aesthetics and remove background grid lines
sns.set(style="whitegrid")
sns.set_theme(style="whitegrid", rc={"grid.linestyle": ''})

# Create subplots for bar plot and line plot
fig, ax = plt.subplots(figsize=(12, 6))

# Bar plot
sns.barplot(x='Model', y='Total Accuracy (Mean)', data=df, color='#C4A5DE', ax=ax)
plt.ylabel('Total Accuracy (Mean)')

# Line plot
sns.lineplot(x='Model', y='Total Accuracy (Mean)', data=df, marker='o', color='red', ax=ax)
plt.ylabel('Total Accuracy (Mean)')

# Annotate the bar plot with specific values
for i in range(len(df)):
    ax.text(i, df['Total Accuracy (Mean)'].iloc[i] + 0.5, f"{df['Total Accuracy (Mean)'].iloc[i]:.2f}",
            ha='center', va='bottom', fontsize=8)

# Set the title
plt.title('Single Client (Hospital 18)')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

# compare


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data for Centralized Learning
centralized_data = {
    'Model': ['VGG 19', 'ResNet 50', 'ResNet 101', 'ResNet 152', 'DenseNet 161'],
    'Total Accuracy (Mean)': [89.3554, 88.3683, 89.291, 91.9447, 82.4314],
    'Total Accuracy (Std Dev)': [1.3163, 1.1088, 1.3979, 1.2797, 1.5482],
}

# Data for Federated Learning
federated_data = {
    'Model': ['VGG 19', 'ResNet 50', 'ResNet 101', 'ResNet 152', 'DenseNet 161'],
    'Total Accuracy (Mean)': [88.1745, 87.4984, 89.0904, 91.1242, 83.8643],
    'Total Accuracy (Std Dev)': [1.4956, 1.4084, 1.3926, 1.3936, 2.198],
}

# Create DataFrames
df_centralized = pd.DataFrame(centralized_data)
df_federated = pd.DataFrame(federated_data)

# Set a color palette for better distinction
colors = sns.color_palette("Set2", n_colors=2)

# Create subplots for line plot
fig, ax = plt.subplots(figsize=(12, 6))

# Line plot for Centralized Learning
sns.lineplot(x='Model', y='Total Accuracy (Mean)', data=df_centralized, marker='o', label='Centralized Learning', color='#aec7e8', linewidth=1)

# Line plot for Federated Learning
sns.lineplot(x='Model', y='Total Accuracy (Mean)', data=df_federated, marker='o', label='Federated Learning', color='#1f77b4', linewidth=2.5)

# Annotate points with numerical values and increased spacing
for i, row in df_centralized.iterrows():
    plt.text(i, row['Total Accuracy (Mean)'] + 0.1, f"{row['Total Accuracy (Mean)']:.2f}", ha='center', va='bottom', fontsize=10, color='#aec7e8')

for i, row in df_federated.iterrows():
    plt.text(i, row['Total Accuracy (Mean)'] - 0.5, f"{row['Total Accuracy (Mean)']:.2f}", ha='center', va='top', fontsize=10, color='#1f77b4')

# Set the title and labels
plt.title('Total Accuracy Comparison: Centralized Learning vs Federated Learning')
plt.xlabel('Model')
plt.ylabel('Total Accuracy (Mean)')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=0)

# Show legend
plt.legend()

# Show the plot
plt.show()
```
