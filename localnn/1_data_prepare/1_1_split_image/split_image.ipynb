# 1 split image by hospital

Folder stracture:

dataset
├── AFF
│   ├── hospital_1
│   │   ├── patient_BCEXMHRRAZ
│   │   │   └── image1.png

├── NFF
│   ├── hospital_1
│   │   ├── ...
│   ├── hospital_2
│   │   ├── ...

NB: some hospitals only have one type images, either appear in AFF or NFF folder.

split by parts of image names.
for example:

patient_ABEFSLVKVT_hospital_55_NFF_image_1

[1]: patient id

[3]: hospital number id

[4]: image type

```python
import os
import shutil

# down-sampled images
image_folder = '/Users/hong/thesis/code/data/images'

# new dataset folder
dataset_folder = '/Users/hong/thesis/code/data/hospital'

# create it if not exist
if not os.path.exists(dataset_folder):
    os.makedirs(dataset_folder)

# AFF and NFF folders
fracture_types = ['AFF', 'NFF']

# read images 
for filename in os.listdir(image_folder):
        parts = filename.split('_')
        image_type = parts[4]
        hospital_number = parts[3]
        patient_id = parts[1]

        # fracture types
        if image_type == 'AFF':
            fracture_types = 'AFF'
        elif image_type == 'NFF':
            fracture_types = 'NFF'
        else:
            continue
    
        # fracture type folders
        fracture_types_folder = os.path.join(dataset_folder, fracture_types)
        os.makedirs(fracture_types_folder, exist_ok=True)

        # hospital-specific folders
        hospital_folder = os.path.join(fracture_types_folder, f'hospital_{hospital_number}')
        os.makedirs(hospital_folder, exist_ok=True)

        # patient-specific folders
        patient_folder = os.path.join(hospital_folder, f'patient_{patient_id}')
        os.makedirs(patient_folder, exist_ok=True)

        # Move the image to the patient-specific folders
        src_path = os.path.join(image_folder, filename)
        dest_path = os.path.join(patient_folder, filename)
        # shutil.move(src_path, dest_path)
        shutil.copy(src_path, dest_path)

```

# 3 create train, validation and test folder by patient

```python
import os
import random
import shutil

# 原始数据集根文件夹
dataset_root = '/Users/hong/thesis/code/data/patient'

# 目标数据集文件夹
target_dataset_folder = '/Users/hong/thesis/code/data/Data'

# 创建目标数据集文件夹
os.makedirs(target_dataset_folder, exist_ok=True)

# AFF和NFF文件夹
fracture_types = ['AFF', 'NFF']

# 分割数据集的比例
train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1

# 遍历AFF和NFF文件夹
for fracture_type in fracture_types:
    source_folder = os.path.join(dataset_root, fracture_type)
    patients = os.listdir(source_folder)
    
    # 随机打乱患者顺序
    random.shuffle(patients)
    
    # 划分数据集
    total_patients = len(patients)
    train_end = int(total_patients * train_ratio)
    val_end = train_end + int(total_patients * val_ratio)
    
    train_patients = patients[:train_end]
    val_patients = patients[train_end:val_end]
    test_patients = patients[val_end:]
    
    # 复制图片到相应的数据集文件夹
    for patient_id in train_patients:
        source_path = os.path.join(source_folder, patient_id)
        target_path = os.path.join(target_dataset_folder, 'train', fracture_type, patient_id)
        shutil.copytree(source_path, target_path)
    
    for patient_id in val_patients:
        source_path = os.path.join(source_folder, patient_id)
        target_path = os.path.join(target_dataset_folder, 'val', fracture_type, patient_id)
        shutil.copytree(source_path, target_path)
    
    for patient_id in test_patients:
        source_path = os.path.join(source_folder, patient_id)
        target_path = os.path.join(target_dataset_folder, 'test', fracture_type, patient_id)
        shutil.copytree(source_path, target_path)

print("Dataset split completed.")
```

```python
import os

def count_number(path_folder):
    # AFF和NFF文件夹路径
    aff_folder = os.path.join(path_folder, 'AFF')
    nff_folder = os.path.join(path_folder, 'NFF')

    # 统计AFF和NFF文件夹中的图片数量
    aff_image_count = sum(len(files) for _, _, files in os.walk(aff_folder))
    nff_image_count = sum(len(files) for _, _, files in os.walk(nff_folder))
    
    return aff_image_count, nff_image_count

print('total:')
path_folder = '/Users/hong/thesis/code/data/patient'
total_aff_num, total_nff_num = count_number(path_folder)
print('aff number:', total_aff_num, '; nff number:', total_nff_num)

print('train:')
path_folder = '/Users/hong/thesis/code/data/Data/train'
aff_num, nff_num = count_number(path_folder)
print('aff number:', aff_num, '; nff number:', nff_num)
print('%:', aff_num / total_aff_num , '; %:', nff_num / total_nff_num)

print('val:')
path_folder = '/Users/hong/thesis/code/data/Data/val'
aff_num, nff_num = count_number(path_folder)
print('aff number:', aff_num, '; nff number:', nff_num)
print('%:', aff_num / total_aff_num , '; %:', nff_num / total_nff_num)


print('test:')
path_folder = '/Users/hong/thesis/code/data/Data/test'
aff_num, nff_num = count_number(path_folder)
print('aff number:', aff_num, '; nff number:', nff_num)
print('%:', aff_num / total_aff_num , '; %:', nff_num / total_nff_num)
```
